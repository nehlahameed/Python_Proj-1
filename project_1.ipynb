```python
import pandas as pd
```

```python
import numpy as np
```

```python
data=pd.read_csv("Python_proj_dataset.csv")
data
```

```python
total=np.random.randint(150,180)
data["Height"].fillna(total,inplace=True)
data
```

#### 1.How many are there in each Team and the percentage splitting with respect to the total employees.

```python
team_counts = data['Team'].value_counts()
percentage_split = (team_counts / len(data)) * 100
print(team_counts)
print(percentage_split)
```

#### 2.Segregate the employees w.r.t different positions.

```python
position_counts = data['Position'].value_counts()
print(position_counts)
```

#### 3.Find from which age group most of the employees belong to.

```python
age_counts = data['Age'].value_counts()
most_common = age_counts.idxmax()
print(most_common)
```

#### 4.Find out under which team and position, spending in terms of salary is high.

```python
high_salary_team_position = data.groupby(['Team', 'Position'])['Salary'].sum().idxmax()
print(high_salary_team_position)
```

#### 5.Find if there is any correlation between age and salary , represent it visually.

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.scatterplot(x='Age', y='Salary', data=data)
plt.title('Correlation Between Age and Salary')
plt.show()
```
